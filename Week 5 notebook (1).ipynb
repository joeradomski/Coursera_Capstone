```python
df_chin = df[df['categories']=='Chinese Restaurant']
df_ind = df[df['categories']=='Indian Restaurant']
df_asian = df[df['categories']=='Asian Restaurant']
df_japan = df[df['categories']=='Japanese Restaurant']
df_sea = df[df['categories']=='Seafood Restaurant']
df_kor= df[df['categories']=='Korean Restaurant']
df_thai = df[df['categories']=='Thai Restaurant']
df_veg = df[df['categories']=='Vegetarian / Vegan Restaurant']
df_ital = df[df['categories']=='Italian Restaurant']
```

```python
venues_map = folium.Map(location=[1.3408528, 103.878446863736], zoom_start=13) # generate map centred around the singapore coordinates
```

```python
for lat, lng, label in zip(df_chin.lat, df_chin.lng, df_chin.categories):
    folium.features.CircleMarker(
        [lat, lng],
        radius=1,
        color= 'blue',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(venues_map)
    
for lat, lng, label in zip(df_ind.lat, df_ind.lng, df_ind.categories):
    folium.features.CircleMarker(
        [lat, lng],
        radius=1,
        color= 'red',
        popup=label,
        fill = True,
        fill_color='red',
        fill_opacity=0.6
    ).add_to(venues_map)
    
for lat, lng, label in zip(df_ital.lat, df_ital.lng, df_ital.categories):
    folium.features.CircleMarker(
        [lat, lng],
        radius=1,
        color= 'green',
        popup=label,
        fill = True,
        fill_color='green',
        fill_opacity=0.6
    ).add_to(venues_map)
    
for lat, lng, label in zip(df_thai.lat, df_thai.lng, df_thai.categories):
    folium.features.CircleMarker(
        [lat, lng],
        radius=1,
        color= 'orange',
        popup=label,
        fill = True,
        fill_color='orange',
        fill_opacity=0.6
    ).add_to(venues_map)
# display map
venues_map
```
